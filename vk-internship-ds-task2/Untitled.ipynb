```python
import pandas as pd
import numpy as np
```

```python
# Пример данных
data = {
    'id': [19114] * 80000,
    'dates': [pd.date_range(start='1/1/2016', periods=60, freq='MS')] * 80000,
    'values': [[-1.86, 0.79, 1.4, 0.15, 0.0, -1.24, -1.46, 3.1] * 7 + [0.5]] * 80000,
    'label': np.random.choice([0.0, 1.0], size=80000)
}
df = pd.DataFrame(data)

# Преобразуем даты в названия фичей (столбцов)
date_features = [f'date_{i}' for i in range(60)]

# Проверка наличия всех значений и добавление NaN, если их меньше 60
def pad_values(values, length=60):
    if len(values) < length:
        values = np.pad(values, (0, length - len(values)), 'constant', constant_values=np.nan)
    return values

df['values'] = df['values'].apply(pad_values)

# Добавляем значения как фичи
for i in range(60):
    df[f'value_{i}'] = df['values'].apply(lambda x, i=i: x[i])

# Формируем итоговый DataFrame
feature_cols = [f'value_{i}' for i in range(60)]
X = df[feature_cols]
y = df['label']
```
